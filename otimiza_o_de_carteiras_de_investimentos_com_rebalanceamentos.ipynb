# **Otimização Multiobjetivo de Carteiras de Investimentos com Rebalanceamento Utilizando NSGA-II e Modelos de Sobrevivência Baseados em Weibull com Frailty**


Este trabalho propõe uma abordagem integrada para a otimização de carteiras de investimentos, considerando simultaneamente três critérios fundamentais: retorno esperado, risco associado e custo operacional decorrente do rebalanceamento da carteira. Partindo da premissa de que as decisões financeiras devem incorporar múltiplas dimensões de análise e refletir a dinâmica temporal dos ativos, a metodologia combina técnicas de modelagem estatística, análise fundamentalista e algoritmos evolutivos de otimização.

A modelagem do risco é realizada por meio de Modelos de Sobrevivência com distribuição de Weibull, incorporando o conceito de frailty, que representa fontes de heterogeneidade não observáveis entre os ativos. Essa abordagem permite estimar o tempo até a ocorrência de eventos relevantes de mercado — como quedas bruscas ou valorizações acentuadas — levando em consideração tanto fatores explicativos conhecidos quanto influências latentes que impactam o comportamento individual de cada ativo.

A estimação dos parâmetros dos modelos de sobrevivência é conduzida via método da Máxima Verossimilhança, garantindo rigor estatístico na adaptação aos dados históricos. As informações utilizadas provêm de três fontes principais: (i) simulações de relatórios de composição de carteiras que refletem aportes reais; (ii) cotações e dados do mercado acionário extraídos da API da Brapi; e (iii) indicadores de análise fundamentalista, que fornecem uma base quantitativa e qualitativa para avaliação dos ativos.

Com os riscos modelados e os ativos avaliados, é realizada a identificação dos pontos ótimos de rebalanceamento utilizando Otimização Bayesiana, considerando variáveis como aporte, preço atual, preço médio, número de ações e a pontuação fundamentalista. Em seguida, aplica-se uma estratégia de otimização multiobjetivo com o algoritmo evolutivo NSGA-II (Non-dominated Sorting Genetic Algorithm II), que gera uma Fronteira de Pareto com soluções não dominadas. Cada ponto dessa fronteira representa uma configuração distinta de carteira, equilibrando os três objetivos simultaneamente: maximização do retorno, minimização do risco e redução dos custos de rebalanceamento.

A visualização tridimensional das soluções eficientes permite identificar perfis estratégicos distintos, como o ponto de maior retorno, o de menor risco ou o de equilíbrio ótimo entre os objetivos. Essa análise é essencial para apoiar a tomada de decisão em ambientes de incerteza, fornecendo ao investidor uma ferramenta robusta para alocação de ativos conforme seu perfil de risco e metas financeiras. Cabe destacar que o foco do estudo é exclusivamente na aplicação do NSGA-II, não sendo realizadas comparações com outros algoritmos evolutivos, de modo a validar a viabilidade e consistência da abordagem proposta.

# **Avaliador de Ações com base em Fundamentos**

Este script Python realiza uma análise fundamentalista de ações da B3 utilizando os dados do site Fundamentus. O objetivo é pontuar e ranquear ações com base em critérios quantitativos de qualidade e valuation.

⚙️ Funcionalidades
Conecta-se automaticamente ao Fundamentus e carrega os dados financeiros das empresas listadas na bolsa.

Aplica critérios básicos e avançados para avaliar a qualidade da empresa, tais como:

Dividend Yield

Preço/Lucro (P/L)

ROE, ROIC, Margem Líquida

Endividamento, Liquidez, Crescimento de Receita

Para cada critério atendido, a empresa recebe 1 ponto.

Ao final, exibe a pontuação de cada empresa e gera um ranking comparativo.

⚠️ Penalidade
Se uma ação apresentar Dividend Yield abaixo de 6%, ela recebe uma penalização automática, tendo sua pontuação total reduzida em 50%.

```python
# --- Caminho do OneDrive (ajuste se necessário) ---
caminho_onedrive = os.path.expanduser('/content/drive/MyDrive/datasets/dados-acoes')
dados_fundamentalistas = 'pontuacao-dados-fundamentalistas.csv'
caminho_completo = os.path.join(caminho_onedrive, dados_fundamentalistas)
```

```python
# --- Definição dos critérios e regras (pontuação 1 ponto por critério atendido) ---
CRITERIOS_BASICOS = {
    'Div.Yield': lambda x: x >= 0.06,                # Preferência por empresas com bons dividendos (>6%)
    'P/L': lambda x: 0 < x <= 15,                    # Evita empresas muito caras em relação ao lucro
    'ROE': lambda x: x >= 0.10,                      # Retorno sobre patrimônio líquido decente
    'Dív.Líq/ Patrim.': lambda x: x < 1,            # Empresas pouco endividadas
    'Cresc. Rec.5a': lambda x: x >= 0.05             # Crescimento estável de receitas
}
```

```python
# --- Critérios avançados ---
CRITERIOS_AVANCADOS = {
    'P/VP': lambda x: x < 2,                         # Preço/Valor Patrimonial baixo
    'EV/EBITDA': lambda x: x < 10,                   # Valuation ajustado pela dívida
    'Mrg. Líq.': lambda x: x > 0.10,                 # Margem líquida acima de 10%
    'ROIC': lambda x: x > 0.10,                      # Retorno sobre capital investido
    'Liq. Corr.': lambda x: x > 1.5,                 # Liquidez corrente acima de 1.5
    'Liq.2meses': lambda x: x > 1000000              # Liquidez de mercado razoável (> R$1M)
}
```

```python
# --- Setores priorizados (com base nos códigos 1, 2, 3, 5, 32, 33, 35, 38, 42) ---
SETORES_PRIORIZADOS = {
    'Agropecuária',
    'Água e Saneamento',
    'Alimentos Processados',
    'Material de Transporte',
    'Energia Elétrica',
    'Máquinas e Equipamentos',
    'Previdência e Seguros',
    'Automóveis e Motocicletas'
    'Químicos',
    'Construção Civil',
    'Telecomunicações',
    'Serv.Méd.Hospit. Análises e Diagnósticos',
    'Tecidos, Vestuário e Calçados',
    'Mineração',
    'Madeira e Papel',
    'Intermediários Financeiros',
    'Petróleo, Gás e Biocombustíveis',
}
```

```python
# --- Combina os dois grupos de critérios ---
CRITERIOS = {**CRITERIOS_BASICOS, **CRITERIOS_AVANCADOS}
```

```python
# --- Carrega os dados do Fundamentus ---
def carregar_dados_fundamentus():
    print("⏳ Carregando dados do Fundamentus...")
    df = fundamentus.get_resultado_raw()
    print("✅ Dados carregados com sucesso!")
    return df
```

```python
# --- Formata valores numéricos e percentuais ---
def formatar_valor(valor, percentual=False):
    if percentual:
        return f"{valor * 100:.1f}%"
    elif valor >= 1000:
        return f"R${valor:,.0f}".replace(",", ".")
    return f"{valor:.2f}"
```

```python
# --- Avalia uma ação com base nos critérios definidos ---
def avaliar_acao(ticker, df):
    print(f"\n🔍 Analisando {ticker}...")

    if ticker not in df.index:
        print(f"⚠️  {ticker} não encontrada nos dados.")
        return 0

    dados_acao = df.loc[ticker]
    pontuacao = 0
    total_criterios = len(CRITERIOS)

    for criterio, regra in CRITERIOS.items():
        try:
            valor_bruto = dados_acao[criterio]

            # --- Tratamento de Conversão (Resolve o erro do seu console) ---
            if isinstance(valor_bruto, str):
                # Remove símbolos comuns e troca vírgula por ponto
                valor = valor_bruto.replace('.', '').replace(',', '.').replace('%', '').strip()
                valor = float(valor)
                # Se era um percentual (ex: 10,50%), divide por 100
                if '%' in valor_bruto:
                    valor = valor / 100
            else:
                valor = float(valor_bruto)

            if pd.isna(valor):
                print(f"  ⚠️  '{criterio}' é NaN. Ignorando.")
                continue

            # --- Formatação para o Log ---
            criterios_percentuais = ['Div.Yield', 'ROE', 'Cresc. Rec.5a', 'Mrg. Líq.', 'ROIC']
            valor_formatado = formatar_valor(valor, percentual=criterio in criterios_percentuais)

            # --- Aplicação da Regra ---
            if regra(valor):
                pontuacao += 1
                print(f"  ✅ {criterio}: {valor_formatado} (Critério atendido)")
            else:
                print(f"  ❌ {criterio}: {valor_formatado} (Critério não atendido)")

        except (ValueError, TypeError):
            print(f"  ❌ Erro de conversão em '{criterio}': valor recebido '{valor_bruto}'")
        except KeyError:
            print(f"  ⚠️  Coluna '{criterio}' não encontrada.")

    # # --- Verifica setor priorizado ---
    try:
        df_papel = fundamentus.get_papel(ticker)
        setor = df_papel.loc[ticker, 'Setor']

        if setor in SETORES_PRIORIZADOS:
            pontuacao += 1
            print(f"  ⭐ Setor '{setor}' é priorizado! (+1 ponto extra)")
        else:
            print(f"  ℹ️  Setor '{setor}' não está entre os priorizados.")

    except Exception as e:
        print(f"  ⚠️  Erro ao verificar setor de {ticker}: {e}")

    # --- Penalização por Div.Yield baixo (<6%) ---
    try:
        dy = dados_acao['Div.Yield']
        if not pd.isna(dy) and dy < 0.06:
            pontuacao_original = pontuacao
            pontuacao = max(0, pontuacao * 0.5)
            print(f"  ⚠️ Dividend Yield abaixo de 6%: {dy:.2%} → Penalização aplicada (50% da pontuação)")
            print(f"     Pontuação ajustada de {pontuacao_original:.1f} para {pontuacao:.1f}")
    except Exception as e:
        print(f"  ⚠️ Erro ao verificar Div.Yield para penalização: {e}")

    percentual_total = (pontuacao / (total_criterios + 1)) * 100
    print(f"📊 Pontuação final para {ticker}: {pontuacao:.1f}/{total_criterios + 1} ({percentual_total:.2f}%)")

    return percentual_total
```

```python
# --- Execução principal ---
def executar_avaliacao(lista_acoes):
    df = carregar_dados_fundamentus()
    resultados = []

    for acao in lista_acoes:
        pontuacao = avaliar_acao(acao, df)
        resultados.append((acao, pontuacao))

    print("\n📋 Ranking Final:")
    df_resultado = pd.DataFrame(resultados, columns=["Ação", "Pontuação"]).sort_values(by="Pontuação", ascending=False)
    df_resultado['Pontuação'] = df_resultado['Pontuação'].map(lambda x: f"{x:.2f}")

    print(df_resultado)

    # --- Salvando CSV ---
    try:
        df_resultado.to_csv(caminho_completo, index=False, sep=';', encoding='utf-8-sig')
        print(f"\n💾 Arquivo CSV salvo com sucesso em: {caminho_completo}")
    except Exception as e:
        print(f"❌ Erro ao salvar o arquivo CSV: {e}")

    return df_resultado
```

# **Pré-processamento e Organização dos Dados Operacionais**

Os dados utilizados na análise foram inicialmente convertidos para o formato adequado, sendo a coluna de datas transformada para o tipo datetime com o intuito de permitir cálculos temporais precisos. A identificação dos ativos foi realizada a partir da coluna textual que combinava o código do ativo e sua descrição. Para isso, foi extraído o ticker com base na estrutura típica desses códigos, e posteriormente padronizado para conter apenas o código do ativo, eliminando qualquer descrição adicional.

As operações de compra e venda foram mapeadas para valores binários, de forma que as compras representassem a entrada do ativo na carteira (valor 1) e as vendas indicassem sua saída (valor 0). Essa conversão possibilitou a utilização de técnicas de modelagem de sobrevivência, nas quais eventos binários são fundamentais para descrever o tempo de permanência de um ativo na carteira. O tempo decorrido desde cada operação até a data de referência (data atual) foi calculado em dias, garantindo que valores nulos ou iguais a zero fossem evitados por meio da adição de uma constante.

Para assegurar a correta manipulação dos valores monetários, os dados referentes ao valor total da operação e ao preço unitário de cada ativo foram convertidos de strings formatadas com símbolos de moeda e separadores decimais para valores numéricos no formato float. Essa transformação foi essencial para a realização de cálculos estatísticos e financeiros posteriores.

Por fim, foi construída uma estrutura em formato de série temporal, por meio do agrupamento dos dados por data e ticker. A média do preço unitário dos ativos foi calculada para cada data, organizando os resultados em uma matriz onde cada coluna representa um ativo e cada linha corresponde a uma data específica. Essa matriz de preços históricos constitui a base para a estimação de modelos de distribuição de sobrevivência, bem como para a aplicação dos algoritmos de otimização de alocação de ativos.

# **Cálculo do Fator de Frailty com Base na Variação Percentual e Pontuação Fundamentalista**

Com o objetivo de incorporar heterogeneidade não observável aos modelos de análise de sobrevivência aplicados ao comportamento de ativos financeiros, implementou-se o cálculo de um fator de frailty (fragilidade) que considera tanto a performance de mercado quanto aspectos fundamentalistas de cada empresa. O conceito de frailty, amplamente utilizado em modelos estatísticos como o de Cox proporcional com efeitos aleatórios, representa fatores latentes que afetam o risco de um evento — como a saída de um ativo da carteira — ocorrer ao longo do tempo.

O procedimento teve início com a definição de três constantes: um fator de ajuste baseado em fundamentos (limitado a 50% de influência), uma penalização adicional de 60% aplicada a empresas com pontuação fundamentalista inferior a 70, e uma pontuação padrão de 50 atribuída a ativos sem informações disponíveis. Em seguida, foi criado um dicionário associando os tickers à sua respectiva pontuação fundamentalista, obtida previamente a partir de uma base estruturada de indicadores contábeis.

Para cada ativo na lista `dados_ativos`, calculou-se o fator de frailty segundo a fórmula:

$$
\text{frailty} = (1 + \text{variação percentual}) \times \left(1 - \frac{\text{pontuação}}{100} \times \text{ajuste fundamentalista} \right)
$$

Nos casos em que a pontuação fundamentalista do ativo era inferior a 70, o valor resultante foi multiplicado por um fator de penalização adicional, representando o risco ampliado de empresas com fundamentos frágeis. Para ativos com dados ausentes, atribuiu-se um valor neutro de frailty igual a 1.0.

Os fatores calculados foram armazenados no dicionário frailty_fatores e posteriormente mapeados ao DataFrame principal df por meio do método map, criando uma nova coluna denominada Frailty. A ausência de dados para determinado ticker foi tratada com imputação padrão (1.0), de forma a não introduzir viés indevido.

Para as análises de sobrevivência, foi construído um novo DataFrame denominado df_eventos, contendo as colunas essenciais: o identificador do ativo (Ativo), o tempo em dias até o evento (Tempo), o tipo de evento (Evento, com 1 para compra e 0 para venda) e o respectivo fator de frailty. Esta estrutura de dados está apta a ser utilizada em modelos estatísticos avançados que incorporam efeitos aleatórios, permitindo a modelagem do risco de forma mais realista e personalizada de acordo com as características de cada ativo.

```python
# --------------------------------------------------------------------------------------
# 1. CALCULA FRAILTY BASEADO NA VARIAÇÃO PERCENTUAL E PONTUAÇAO FUNDAMENTALISTA
# --------------------------------------------------------------------------------------

# Constantes de configuração
AJUSTE_FUNDAMENTALISTA = 0.5    # Até 50% de ajuste no frailty baseado na pontuação
PENALIZACAO_EXTRA = 1.6         # Penalização adicional de 60% se pontuação < 70
PONTUACAO_PADRAO = 50.0         # Pontuação padrão se não houver dados na base

# Cria dicionário: {'VALE3': 83.33, ...}
pontuacoes_fundamentalistas = df_fund.set_index('Ação')['Pontuação'].to_dict()

frailty_fatores = {}

for dados in dados_ativos:
    ticker = dados['Ticker']
    variacao_bruta = dados.get('Crescimento 5 anos %') # Nome usado no seu dicionário anterior

    # Busca a pontuação (garantindo que seja float)
    pontuacao_raw = pontuacoes_fundamentalistas.get(ticker, PONTUACAO_PADRAO)
    try:
        pontuacao = float(pontuacao_raw)
    except:
        pontuacao = PONTUACAO_PADRAO

    # --- TRATAMENTO DA VARIAÇÃO (Conversão de String para Float) ---
    variacao = 0.0
    try:
        if isinstance(variacao_bruta, str):
            # Remove pontos de milhar, troca vírgula por ponto e tira o símbolo %
            clean_val = variacao_bruta.replace('.', '').replace(',', '.').replace('%', '').strip()
            variacao = float(clean_val)
            # Se era percentual no texto (ex: "39.7%"), converte para decimal (0.397)
            if '%' in variacao_bruta:
                variacao /= 100
        elif variacao_bruta is not None:
            variacao = float(variacao_bruta)
    except (ValueError, TypeError):
        variacao = 0.0 # Valor neutro em caso de erro catastrófico na conversão

    # --- CÁLCULO DA FRAILTY ---
    # Validação dos dados para o cálculo
    if pd.notnull(variacao) and pd.notnull(pontuacao):
        # Cálculo da frailty ajustada
        # (1 + variacao) reflete o crescimento, (1 - ajuste) reflete a qualidade fundamentalista
        frailty = (1 + variacao) * (1 - (pontuacao / 100) * AJUSTE_FUNDAMENTALISTA)

        # Penalização adicional para ativos com pontuação menor que 70
        if pontuacao < 70.0:
            frailty *= PENALIZACAO_EXTRA
    else:
        frailty = 1.0

    frailty_fatores[ticker] = frailty

    # Log de acompanhamento (opcional, pode remover)
    print(f'{ticker:<6} - Frailty: {frailty:.2f} - Pontuação: {pontuacao:.2f} - Variação: {variacao if variacao is not None else "N/A"}')

# Validação das colunas mínimas
required_cols = {'Ticker', 'Tempo', 'Evento'}
if not required_cols.issubset(df.columns):
    raise ValueError(f"O DataFrame está faltando as colunas: {required_cols - set(df.columns)}")

# Aqui usamos apenas os tickers com dados válidos
ativos = [dado.get('Ticker') for dado in dados_ativos if dado.get('Ticker') is not None]

# Aplica o frailty ao DataFrame
df['Frailty'] = df['Ticker'].map(frailty_fatores).fillna(1.0)

# Cria DataFrame de eventos final
df_eventos = df[['Ticker', 'Tempo', 'Evento', 'Frailty']].copy()
df_eventos = df_eventos.rename(columns={'Ticker': 'Ativo'})
```

# **Ajuste do Modelo Weibull AFT com Frailty via Variáveis Dummies**

Com o objetivo de modelar o tempo de permanência dos ativos na carteira sob a ótica da análise de sobrevivência, foi utilizado o modelo AFT (Accelerated Failure Time) com distribuição de Weibull. O modelo AFT é uma alternativa ao modelo de riscos proporcionais de Cox, sendo particularmente adequado quando se deseja interpretar diretamente o efeito das covariáveis sobre o tempo até a ocorrência do evento. Neste contexto, o evento de interesse é a venda do ativo, e o tempo é o número de dias entre a compra e a venda.

Para capturar os efeitos específicos de cada ativo sobre o tempo de sobrevivência, foram utilizadas variáveis dummies, geradas a partir da coluna categórica `Ativo`. Isso foi feito por meio da função pd.get_dummies(), com o parâmetro drop_first=False, de forma que todas as categorias foram mantidas — uma decisão importante, pois permite estimar os efeitos absolutos de cada ativo, e não apenas em relação a uma categoria de referência.

Além das variáveis indicadoras dos ativos, também foi incluída a coluna `Frailty`, calculada na etapa anterior. Esta variável contínua representa uma tentativa de capturar a heterogeneidade não observada entre os ativos, funcionando como um fator de ajuste adicional no modelo de sobrevivência.

A fórmula utilizada para calcular o frailty de cada ativo foi:


$$
\text{frailty} = (1 + \text{variação percentual}) \times \left(1 - \frac{\text{pontuação}}{100} \times \text{ajuste fundamentalista} \right)
$$

Essa métrica visa combinar informações de retorno recente (variação percentual) e fundamentos econômicos (ajuste fundamentalista ponderado pela pontuação do ativo), compondo um índice sintético que reflete a atratividade ou vulnerabilidade implícita de cada ativo.

Com a estrutura do DataFrame `df_model` preparada, foi utilizado o estimador `WeibullAFTFitter()` da biblioteca lifelines para ajustar o modelo, com `Tempo` como variável de duração e `Evento` como indicador de censura. Após o ajuste, foram extraídos os coeficientes associados à componente de escala (`lambda_`), que representam a contribuição de cada ativo para o tempo de permanência estimado.

Esses coeficientes foram interpretados como estimativas do frailty individual de cada ativo, revelando quais ativos tendem a permanecer mais ou menos tempo na carteira, mesmo após o controle da variação percentual e do ajuste fundamentalista.

Essa análise fornece informações valiosas sobre o comportamento típico dos investidores frente a diferentes ativos e pode ser utilizada para embasar decisões relacionadas a **estratégias de rotação, permanência e diversificação da carteira.**

```python
# -----------------------------
# 2. AJUSTE DO MODELO WEIBULL AFT COM FRAILTY VIA DUMMIES
# -----------------------------

# Cria dummies para as variáveis categóricas dos ativos
df_model = pd.get_dummies(
    df_eventos[['Tempo', 'Evento', 'Ativo', 'Frailty']],
    columns=['Ativo'],
    drop_first=False  # Mantém todas as categorias (não exclui nenhuma referência)
)

# Instancia e ajusta o modelo Weibull AFT
aft = WeibullAFTFitter()

# Ajusta o modelo considerando a duração (Tempo) e o evento (Evento)
aft.fit(df_model, duration_col='Tempo', event_col='Evento')

# EXTRAÇÃO DOS COEFICIENTES (FRAILTY)
# Extrai os coeficientes relacionados à lambda_ (escala do Weibull)
coeficientes = aft.params_.loc['lambda_']

# Cria dicionário com os nomes dos ativos e seus respectivos frailty
frailty_ativos = {k.replace('Ativo_', ''): v for k, v in coeficientes.to_dict().items() if k.startswith('Ativo_')}

# Exibe os frailty dos ativos estimados pelo modelo
for ativo, valor in frailty_ativos.items():
    print(f'{ativo}: Frailty estimado = {valor:.4f}')
```

Após o ajuste do modelo Weibull AFT com variáveis dummies para os ativos e inclusão da variável contínua Frailty, foi extraído o coeficiente associado especificamente a essa covariável contínua.

```python
frailty_manual = coeficientes.get('Frailty', None)
if frailty_manual is not None:
    print(f'\nFrailty manual (baseado em variação + pontuação): {frailty_manual:.4f}')
```

# **Pré-cálculos para a Etapa de Otimização Bayesiana com Optuna**

Antes de iniciar a quarta etapa da otimização bayesiana com Optuna, são realizados pré-cálculos fundamentais relacionados ao risco e retorno dos ativos disponíveis. Esses cálculos têm como objetivo preparar os insumos necessários para avaliar e comparar diferentes combinações de ativos com base em critérios quantitativos.

Primeiramente, são calculados os retornos diários históricos a partir das séries de preços dos ativos. Com base nesses retornos, obtêm-se os retornos médios e os riscos históricos (medidos como desvio padrão dos retornos) de cada ativo. Essas informações são convertidas em dicionários para facilitar o acesso posterior.

Para incorporar diferentes dimensões de risco, o modelo propõe uma combinação entre o risco histórico e o chamado risco estrutural, representado pela magnitude do parâmetro frailty estimado anteriormente no modelo de sobrevivência. Esse parâmetro captura variações não observadas entre os ativos, funcionando como uma medida adicional de risco associada à dinâmica de entrada e saída de ativos da carteira ao longo do tempo.

A combinação dos dois tipos de risco é feita por meio de pesos ajustáveis: um peso alpha (definido como 0.7) para o risco histórico e um peso beta (definido como 0.3) para o risco estrutural. Caso algum ativo não possua dados disponíveis de risco ou frailty, são utilizados valores padrão para garantir robustez no processamento (risco padrão de 0.1 e frailty padrão de 0.0).

Além disso, são extraídos os preços atuais dos ativos para fins de avaliação financeira e cálculo de alocações. A lista final de ativos é então usada como referência para as etapas seguintes da otimização.

Esses pré-cálculos são essenciais para garantir que a função objetivo da otimização bayesiana leve em conta não apenas o retorno esperado, mas também uma avaliação realista e personalizada do risco de cada ativo, resultando em portfólios mais balanceados e consistentes com os padrões históricos e estruturais observados no mercado.

```python
# --------------------------------------------------------------------------
# PRÉ-CÁLCULOS PARA A FUNÇÃO NA 4ª ETAPA DE OTIMIZAÇÃO BAYESIANA COM OPTUNA
# --------------------------------------------------------------------------

# Silenciar avisos de depreciação para o output ficar limpo
warnings.filterwarnings('ignore', category=FutureWarning)

# 1. FILTRAGEM CRÍTICA: Remover ativos que falharam no download
# Verificamos quais colunas têm dados (não são todas NaN)
dados_mercado = dados_mercado.dropna(axis=1, how='all')

# Atualizamos a lista de ativos apenas com o que o Yahoo encontrou
ativos_validos = dados_mercado.columns.str.replace(".SA", "", regex=False).tolist()
dados_mercado.columns = ativos_validos

# 2. Cálculo de Retornos e Riscos apenas para os válidos
# Preenchemos NaNs pontuais para evitar que o pct_change quebre
retornos_diarios = dados_mercado.ffill().pct_change().dropna()
retornos_ativos = retornos_diarios.mean().to_dict()
riscos_ativos = retornos_diarios.std().to_dict()

# Pesos para combinação de riscos
alpha = 0.7  # peso para risco histórico
beta = 0.3   # peso para risco estrutural (frailty)

# Mapear preços atuais dos ativos
precos_ativos = dados_mercado.iloc[-1].to_dict()
ativos = ativos_validos

# Parâmetros padrão para casos com dados faltantes
RISCO_PADRAO = 0.1
FRAILTY_PADRAO = 0.0

# Combinação dos riscos: histórico + estrutural
riscos_ativos_combinados = {
    ativo: alpha * riscos_ativos.get(ativo, RISCO_PADRAO) +
           beta * abs(frailty_ativos.get(ativo, FRAILTY_PADRAO))  # frailty como risco estrutural (magnitude)
    for ativo in ativos
}

print("✅ Pré-cálculos concluídos. O Optuna agora terá dados para processar.")
```

Nesse ponto tentamos responder à seguinte pergunta:

***“Qual seria a melhor alocação percentual nos ativos para maximizar retorno, minimizar risco e evitar rebalanceamentos caros?”***

Nesta etapa, é realizada a otimização da alocação de ativos por meio de Optuna, uma biblioteca de otimização bayesiana eficiente. O objetivo é encontrar a melhor distribuição de capital entre os ativos disponíveis, considerando um valor fixo de aporte e buscando maximizar o retorno esperado da carteira, ao mesmo tempo em que se minimiza o risco total, de forma realista e prática.

A função funcao_objetivo é o núcleo da otimização. Para cada tentativa (ou trial), o Optuna propõe pesos contínuos para os ativos (variando de 0 a 1), que são posteriormente normalizados para que a soma total seja igual a 1. A partir desses pesos, é calculado o valor em reais que seria alocado em cada ativo, proporcionalmente ao valor total de aporte disponível.

Com base no valor alocado e nos preços atuais dos ativos, calcula-se o número de ações inteiras que podem ser compradas de cada ativo, usando arredondamento para baixo (via np.floor). Se a alocação sugerida resultar em nenhuma compra (ou seja, todas as quantidades forem zero), ou se o valor total das ações exceder o aporte, a função retorna um valor de penalidade (float('inf')) que sinaliza ao Optuna que essa solução deve ser descartada.

A carteira válida é então reavaliada com base nos pesos efetivos reais, levando em consideração o valor efetivamente utilizado na compra das ações. Utilizando esses pesos, calcula-se o retorno esperado (como média ponderada dos retornos dos ativos) e o risco total (como média ponderada dos riscos combinados). A função de objetivo retorna o negativo do valor retorno - risco, indicando que desejamos maximizar esse score, embora a otimização esteja configurada para minimização (daí o sinal negativo).

Após a conclusão dos 300 trials definidos, o código imprime os melhores resultados encontrados, incluindo:

* Quantidade de ações de cada ativo que devem ser compradas;

* Valor investido por ativo;

* Valor total do aporte utilizado;

* Retorno esperado da carteira;

* Risco total estimado.

Por fim, há uma verificação adicional para garantir que o valor alocado não ultrapasse o aporte disponível. Se isso ocorrer, o código emite um aviso; caso contrário, confirma que a alocação está dentro dos limites financeiros.

Essa abordagem possibilita encontrar uma alocação ótima de portfólio com restrição de capital, respeitando preços de mercado e evitando alocações irreais, o que a torna altamente aplicável para situações práticas de investimento.

```python
# -----------------------------------
# 4. OTIMIZAÇÃO BAYESIANA COM OPTUNA COM RESTRIÇÃO DE APORTES
# -----------------------------------

def funcao_objetivo(trial):
    peso_ativos = np.array([trial.suggest_float(f"peso_{i}", 0.0, 1.0) for i in range(len(ativos))])
    peso_soma = peso_ativos.sum()
    if peso_soma == 0:
        return float('inf')
    peso_ativos /= peso_soma  # Normaliza os pesos

    valor_alocado = peso_ativos * VALOR_APORTE

    quantidades = []
    for i, ativo in enumerate(ativos):
        preco = precos_ativos.get(ativo)
        if preco is None:
            raise ValueError(f"Preço não encontrado para o ativo {ativo}")
        qtd = np.floor(valor_alocado[i] / preco)
        quantidades.append(qtd)

    quantidades = np.array(quantidades)

    if quantidades.sum() == 0:
        return float('inf')

    valor_utilizado = np.sum(quantidades * np.array([precos_ativos[ativo] for ativo in ativos]))
    if valor_utilizado > VALOR_APORTE:
        return float('inf')

    pesos_efetivos = (quantidades * np.array([precos_ativos[ativo] for ativo in ativos])) / valor_utilizado

    retornos = np.array([retornos_ativos.get(ativo, 0.0) for ativo in ativos])
    riscos = np.array([riscos_ativos_combinados.get(ativo, 0.1) for ativo in ativos])

    retorno_esperado = np.dot(pesos_efetivos, retornos)
    risco_total = np.dot(pesos_efetivos, riscos)

    # Objetivo: maximizar retorno e minimizar risco (via score = retorno - risco)
    return -(retorno_esperado - risco_total)


# Execução com Optuna

study = optuna.create_study(direction="minimize")
study.optimize(funcao_objetivo, n_trials=300)

# Impressão dos melhores resultados

print("\nMelhores alocações considerando o aporte:")

peso_ativos = np.array([study.best_params.get(f"peso_{i}", 0.0) for i in range(len(ativos))])
peso_ativos /= peso_ativos.sum()
valor_alocado = peso_ativos * VALOR_APORTE

quantidades = np.floor(valor_alocado / np.array([precos_ativos[ativo] for ativo in ativos]))
valor_total_utilizado = 0

pesos_efetivos = []

for i, ativo in enumerate(ativos):
    preco = precos_ativos.get(ativo)
    quantidade = quantidades[i]
    if quantidade > 0:
        valor_ativo = quantidade * preco
        valor_total_utilizado += valor_ativo
        pesos_efetivos.append(valor_ativo)
        print(f"{ativo}: {int(quantidade)} ações (Preço: {preco:.2f}, Valor: {valor_ativo:.2f})")
    else:
        pesos_efetivos.append(0)

pesos_efetivos = np.array(pesos_efetivos) / valor_total_utilizado

print(f"\nValor total utilizado do aporte: R$ {valor_total_utilizado:.2f} de R$ {VALOR_APORTE:.2f}")

retorno_final = np.dot(pesos_efetivos, np.array([retornos_ativos[ativo] for ativo in ativos]))
risco_final = np.dot(pesos_efetivos, np.array([riscos_ativos_combinados[ativo] for ativo in ativos]))

print(f"\nRetorno esperado: {retorno_final:.4f}")
print(f"Risco total: {risco_final:.4f}")

if valor_total_utilizado > VALOR_APORTE:
    print("\n⚠️ Atenção: O valor total utilizado ultrapassa o aporte! Corrija a função de alocação.")
else:
    print("\n✅ A alocação está dentro do valor do aporte.")
```

Este trecho de código realiza a conversão de dicionários com métricas dos ativos em arrays NumPy, mantendo a mesma ordem definida na lista ativos. Essa uniformidade é essencial para garantir que os cálculos vetoriais subsequentes — como multiplicações ponderadas ou otimizações — sejam consistentes e corretos em relação aos ativos.

Primeiramente, o vetor retorno_ativos_array é construído a partir dos valores médios de retorno diário de cada ativo. Em seguida, o vetor risco_ativos_array armazena os riscos combinados (histórico + estrutural), aplicando um valor padrão de risco (0.1) caso algum ativo não tenha risco definido no dicionário riscos_ativos.

Por fim, o vetor rebal_freq_array representa a frequência estimada de rebalanciamento para cada ativo. Essa frequência é calculada como o inverso do tempo mediano de sobrevivência (1 / tempo), com o valor 0 sendo atribuído caso o tempo seja zero, para evitar divisões indefinidas. Esse vetor é derivado do dicionário tempos_por_ticker, que contém o tempo mediano estimado até que cada ativo entre na carteira, com um valor padrão de 1 caso o ativo não esteja presente no dicionário.

Esse conjunto de vetores padronizados é útil para análises quantitativas e otimizações de portfólio que considerem não apenas retorno e risco, mas também a dinâmica de entrada dos ativos na carteira ao longo do tempo, representada pela frequência de rebalanciamento.

```python
# Converte os dicionários para arrays na mesma ordem de 'ativos'
retorno_ativos_array = np.array([retornos_ativos[ativo] for ativo in ativos])

risco_ativos_array = np.array([riscos_ativos.get(ativo, 0.1) for ativo in ativos])

rebal_freq_array = np.array([
    1 / tempo if tempo != 0 else 0
    for tempo in [tempos_por_ticker.get(ativo, 1) for ativo in ativos]
])# frequência ≈ 1 / tempo mediano
```

# **Otimização Multiobjetivo com NSGA-II (via PyMoo)**

Nesta etapa do trabalho, foi implementado um modelo de otimização multiobjetivo utilizando o algoritmo evolutivo NSGA-II (Non-dominated Sorting Genetic Algorithm II), por meio do pacote pymoo. O objetivo é construir carteiras de investimento eficientes considerando simultaneamente três critérios fundamentais:

Maximizar o retorno esperado

Minimizar o risco da carteira

Minimizar o custo de rebalanceamento

Para isso, foi criada a classe personalizada CarteiraMultiObjetivo, que representa formalmente o problema de otimização a ser resolvido. Cada carteira é modelada como um vetor de pesos, onde cada elemento define a proporção do capital alocada em um ativo.

A função de avaliação (_evaluate) calcula:

O retorno esperado como o somatório ponderado dos retornos médios dos ativos;

O risco como a soma ponderada das volatilidades (riscos individuais dos ativos);

O custo de rebalanceamento, representado como uma penalidade proporcional à frequência estimada de alterações na carteira (com base em dados históricos dos ativos).

Como o pymoo realiza minimização por padrão, o retorno é invertido (multiplicado por -1) para garantir que soluções com maior retorno sejam consideradas melhores. Além disso, foi imposta uma restrição de igualdade, exigindo que a soma dos pesos seja igual a 1 — ou seja, que 100% do capital seja distribuído entre os ativos.

A solução do problema é feita em até 100 gerações, com o algoritmo NSGA-II explorando o espaço de soluções viáveis. O resultado é um conjunto de carteiras não-dominadas (Fronteira de Pareto), nas quais não é possível melhorar um dos objetivos sem piorar outro. Isso permite que o investidor escolha uma alocação mais alinhada ao seu perfil de risco e preferências entre retorno e custo operacional.





```python
# -----------------------------
# 5. OTIMIZAÇÃO MULTIOBJETIVO COM NSGA-II
# -----------------------------

# Define o problema de otimização multiobjetivo
class CarteiraMultiObjetivo(Problem):
    def __init__(self):
        super().__init__(
            n_var=len(ativos),   # número de ativos
            n_obj=3,             # retorno, risco e custo
            n_constr=1,          # soma dos pesos = 1
            xl=0.0,              # limite inferior dos pesos
            xu=1.0               # limite superior dos pesos
        )

    def _evaluate(self, x, out, *args, **kwargs):
        x = np.array(x)

        # Calcula os 3 objetivos
        retorno = np.dot(x, retorno_ativos_array)
        risco = np.dot(x, risco_ativos_array)
        custo = np.dot(x, rebal_freq_array) * 0.01  # penalidade por frequência de rebalanceamento

        # Define objetivos (lembrando: pymoo minimiza → -retorno para maximizar)
        out["F"] = np.column_stack([-retorno, risco, custo])

        # Restrição: soma dos pesos deve ser 1
        out["G"] = np.sum(x, axis=1) - 1

# Instancia o problema
problem = CarteiraMultiObjetivo()
```

## **Avaliação da Convergência via Hipervolume e Identificação da Solução Mais Balanceada**

Nesta etapa, foi realizada uma análise da qualidade das soluções ao longo das gerações produzidas pelo algoritmo NSGA-II, utilizando como métrica o indicador de Hipervolume (HV – Hypervolume Indicator). Além disso, foi implementado um critério para determinar a solução mais balanceada, considerando simultaneamente os três objetivos do problema: retorno (maximizado), risco (minimizado) e custo (minimizado).

**Cálculo do Hipervolume (HV) por geração**

Todas as frentes de Pareto geradas ao longo das gerações foram extraídas do histórico salvo (res_nsga2.history) e empilhadas em uma única matriz all_F, contendo os valores dos objetivos de todas as populações intermediárias. Com base nisso, foi calculado um ponto de referência (ref_point), definido como 10% acima dos piores valores em cada objetivo. Esse ponto é necessário para o cálculo do hipervolume, que quantifica o volume do espaço objetivo dominado pelas soluções não dominadas em relação a esse ponto de referência.

A seguir, o hipervolume foi calculado para cada geração individualmente, gerando uma sequência de valores (hv_values) que descreve a evolução da qualidade da fronteira de Pareto ao longo do processo de otimização.

**Seleção da Solução Mais Balanceada**

Para identificar a solução mais equilibrada entre os objetivos conflitantes, foi utilizado um critério baseado na soma dos ranks individuais de cada objetivo:

A matriz pareto_objectives contém os valores dos três objetivos das soluções finais na fronteira de Pareto (retorno negativo, risco e custo);

Para cada objetivo, as soluções foram ordenadas e receberam uma classificação relativa (rank);

As classificações foram somadas para cada solução, e aquela com a menor soma foi considerada a solução mais balanceada, representando um bom compromisso entre retorno, risco e custo;

O vetor melhor_obj armazena os valores dessa solução, e a geração em que ela surgiu foi localizada percorrendo o histórico de populações armazenadas.

**Visualização Gráfica**

Para interpretar os resultados, foi gerado um gráfico contendo:

A curva de evolução do hipervolume ao longo das gerações, refletindo o progresso do algoritmo em encontrar soluções cada vez melhores;

A marcação da geração em que surgiu a solução mais balanceada, destacada com uma linha vertical vermelha e anotada com seus respectivos valores de retorno, risco e custo (em porcentagens);

Um gráfico de zoom no canto inferior direito do subplot, exibindo as últimas 20 gerações para detalhar a estabilização do hipervolume na fase final da otimização.

Essa visualização fornece uma ferramenta prática e interpretável para avaliar o desempenho do algoritmo, além de facilitar a escolha de soluções robustas e equilibradas com base em múltiplos critérios simultâneos.

```python
# ----------------------------------------
# Cálculo do Hipervolume (HV) por geração
# ----------------------------------------

all_F = np.vstack([gen.pop.get("F") for gen in res_nsga2.history])
ref_point = np.max(all_F, axis=0) * 1.1  # ponto de referência 10% acima do pior valor em cada objetivo

hv = HV(ref_point=ref_point)
hv_values = [hv.do(gen.pop.get("F")) for gen in res_nsga2.history]

# -----------------------------
# Encontrar solução mais balanceada (mínima soma de ranks)
# -----------------------------

pareto_objectives = res_nsga2.F  # [ -retorno, risco, custo ]

rankings = np.zeros_like(pareto_objectives)
for i in range(pareto_objectives.shape[1]):
    rankings[:, i] = np.argsort(np.argsort(pareto_objectives[:, i]))

rank_sums = np.sum(rankings, axis=1)
melhor_idx = np.argmin(rank_sums)
melhor_obj = pareto_objectives[melhor_idx]

# Localizar em qual geração essa solução apareceu
geracao_melhor_balanceada = None
for gen_idx, gen in enumerate(res_nsga2.history):
    F_gen = gen.pop.get("F")
    for f in F_gen:
        if np.allclose(f, melhor_obj, rtol=1e-5, atol=1e-8):
            geracao_melhor_balanceada = gen_idx
            break
    if geracao_melhor_balanceada is not None:
        break

# -----------------------------
# Função reutilizável para plotar
# -----------------------------

def plot_hv(ax, hv_values, title, show_annotation=False, ylim_zero=False):
    ax.plot(np.arange(len(hv_values)), hv_values, label='NSGA-II')

    if ylim_zero:
        ax.set_ylim(bottom=0)

    if show_annotation and geracao_melhor_balanceada is not None:
        # Linha vertical da solução balanceada
        ax.axvline(x=geracao_melhor_balanceada, color='red', linestyle='--', label='Solução balanceada')
        ax.scatter(geracao_melhor_balanceada, hv_values[geracao_melhor_balanceada], color='red', zorder=5)

        # Texto de anotação
        x_text = ax.get_xlim()[1] - 10
        y_text = ax.get_ylim()[1] * 0.95

        ax.annotate(
            f'Ret = {-melhor_obj[0]*1000000:.2f}%\nRisco = {melhor_obj[1]*100000:.2f}%\nCusto = {melhor_obj[2]*100000000:.2f}%\nGeração = {geracao_melhor_balanceada}',
            xy=(geracao_melhor_balanceada, hv_values[geracao_melhor_balanceada]),
            xytext=(-10, -70),  # mais à direita (x) e mais abaixo (y)
            textcoords='offset points',
            arrowprops=dict(arrowstyle='->'),
            fontsize=9,
            bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.9),
            ha='right',  # alinha à direita da caixa de texto
            va='top'     # alinha o topo da caixa com o ponto de origem
        )

        # Inserir gráfico de zoom no canto inferior direito
        ax_inset = inset_axes(ax, width="35%", height="35%", loc="lower right")
        zoom_start = max(0, len(hv_values) - 20)
        zoom_end = len(hv_values)
        ax_inset.plot(np.arange(zoom_start, zoom_end), hv_values[zoom_start:zoom_end], color='blue')
        ax_inset.set_title("Zoom (últimas 20 gerações)", fontsize=9)
        ax_inset.grid(True, linestyle='--', alpha=0.4)
        ax_inset.tick_params(labelsize=8)

    ax.set_title(title)
    ax.set_xlabel("Geração")
    ax.set_ylabel("Qualidade da Fronteira (HV)")
    ax.grid(True, linestyle='--', alpha=0.4)
    ax.legend()

# -----------------------------
# Criar gráfico com subplots lado a lado
# -----------------------------

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Gráfico com anotação e zoom
plot_hv(ax1, hv_values, "Convergência com marcação da solução balanceada", show_annotation=True, ylim_zero=True)

# Gráfico com y iniciando em zero
plot_hv(ax2, hv_values, "Convergência com marcação da solução balanceada (com zoom)", show_annotation=True, ylim_zero=False)

plt.tight_layout()
plt.show()
```

Ponto Ideal: a solução mais equilibrada considerando simultaneamente alto retorno, baixo risco e baixo custo. Este ponto foi identificado calculando-se a métrica retorno - risco - custo, maximizando esse valor como uma forma heurística de representar o melhor trade-off entre os objetivos.

Essa visualização é fundamental para compreender os compromissos (trade-offs) entre os objetivos, oferecendo suporte à tomada de decisão com base nas preferências do investidor.

```python
# -----------------------------
# 6. PLOTAR FRONTEIRA DE PARETO
# -----------------------------

def marcar_pontos(ax, F):
    retorno = -F[:, 0]  # Inverte para maximizar
    risco = F[:, 1]
    custo = F[:, 2]

    # Ponto com maior retorno
    idx_max_ret = np.argmax(retorno)
    ponto_max_ret = (retorno[idx_max_ret], risco[idx_max_ret], custo[idx_max_ret])
    ax.scatter(*ponto_max_ret, c='green', s=80, marker='.', label='Maior Retorno')
    ax.text(*ponto_max_ret, '  Maior Retorno', color='green')

    # Ponto ideal: maior retorno, menor risco e menor custo
    idx_ideal = np.argmax(retorno - risco - custo)
    ponto_ideal = (retorno[idx_ideal], risco[idx_ideal], custo[idx_ideal])
    ax.scatter(*ponto_ideal, c='red', s=80, marker='.', label='Ponto Ideal')
    ax.text(*ponto_ideal, '  Ponto Ideal', color='red')

# -----------------------------
# 6. PLOTAR FRONTEIRA DE PARETO
# -----------------------------
fig = plt.figure(figsize=(15, 8))

gs = fig.add_gridspec(1, 2, width_ratios=[1, 1])

# NSGA-II
ax1 = fig.add_subplot(111, projection='3d')
retorno_nsga2 = -res_nsga2.F[:, 0]
ax1.scatter(retorno_nsga2, res_nsga2.F[:, 1], res_nsga2.F[:, 2], c='blue', label='NSGA-II')
ax1.set_title('NSGA-II Fronteira de Pareto')
ax1.set_xlabel('Retorno', fontsize=12, labelpad=10)
ax1.set_ylabel('Risco', fontsize=12, labelpad=10)
ax1.set_zlabel('Custo', fontsize=12, labelpad=10)

ax2 = fig.add_subplot(gs[1])  # subplot vazio
ax2.axis('off')

ax1.view_init(elev=20, azim=135)
marcar_pontos(ax1, res_nsga2.F)
ax1.grid(True, linestyle='--', alpha=0.4)
ax1.legend()
plt.show()
```
